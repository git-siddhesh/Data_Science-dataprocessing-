```
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('salary.csv')
# salary = b0 + b1*experience

# experience that we are going to use as input of training data
exp=df.iloc[:,0:1].values

# actual salary
sal = df.iloc[:,1:2].values

# now we can visualize this exp and sal data
plt.xlabel("Experience")
plt.ylabel("salary")
plt.scatter(exp,sal)
plt.show()
```

```
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
trained= regr.fit(exp,sal)
# actual salary
```

```
# predicted rysala
plt.xlabel("Experience")
plt.ylabel("salary")
plt.scatter(exp,sal,label="actual data")
plt.plot(exp,trained.predict(exp),label="Predicted")
plt.legend()
plt.show()
```
