```
import requests
dir(requests)

# accesing website from url 
webdata = requests.get('https://php.net')
webdata   # response code of HTTP protocol  - 
          #  200 URL is ok, 302 eans redirect, 404  means url not found, 505 means server is not permitted for webscrapping
  
```

*Extract data from webdata*

```
# Now get actual data of any website
html_data = webdata.text    # html data in text form 
```

*Using beautifulsoup for PARSER*

```
# importing bs4
from bs4 import BeautifulSoup
              # htmldata   parser
soup = BeautifulSoup(html_data,'html5lib')
clean_data = soup.get_text()
clean_data

# Saving data permanently 
with open('mywebdata.txt','w+') as f:
  f.write(clean_data)
```

*now we have just save the data in a file now we will xtract data from there*

```
f = open('mywebdata.txt','r')
mydata=f.read()
f.close()
'''
import time
for i in mydata.split():
  print(i)
  time.sleep(1)
 '''
# now split data into list 
newdata = [i for i in mydata.split()]
newdata
```

```
import matplotlib.pyplot as plt
import nltk

[i for i in dir(nltk) if 'Freq' in i]
```

* make a list of words and their counting to be repeated *

```
nltkdata = nltk.FreqDist(newdata)
nltkdata.plot(20)    # top 20 words with
# graph between words and counts


# but it also contain stopwords 
```

*remove stopwords *

```
from nltk.corpus import stopwords
#stopwords.words("english")
finaldata = [i for i in nltkdata if i.lower() not in stopwords.words('english')]
```

```
nltk.FreqDist(finaldata).plot(20)
```
